```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('punkt')


#불용어(영어)
nltk.download('stopwords')
```

```python
#문장 단위 토큰화
text_en = 'God is Great! I won a lottery.'
text_ko = '애들아 안녕. 나는 얼굴 천재 차은우야! ㅎㅎ'

tokenized_en = sent_tokenize(text_en)
tokenized_ko = sent_tokenize(text_ko)

# 결과 확인
print(tokenized_en)
print(tokenized_ko)
```

```python
#단어 단위 토큰화(1)

#문장 설정
text_en2 = 'Learning Python is very exciting!'
text_ko2 = "오늘은 즐겁고 행복한 목요일 입니다:) "

#단어 토큰화 진행 --> word_tokenize() 사용

tokenized_en2 = word_tokenize(text_en2)
tokenized_ko2 = word_tokenize(text_ko2)

splited_en = text_en2.split()
splited_ko = text_ko2.split()

print(splited_en)
print(splited_ko)

print(tokenized_en2)
print(tokenized_ko2)
```

```python
#단어 단위 토큰화(2)

#문장 : text_en, text_ko
tokenized_en = word_tokenize(text_en)
tokenized_ko = word_tokenize(text_ko)

print(tokenized_en)
print(tokenized_ko)
```

## 한글 텍스트 '형태소' 단위 토큰화 실습

```python
!curl -s https://raw.githubusercontent.com/teddylee777/machine-learning/master/99-Misc/01-Colab/mecab-colab.sh | bash
```

```python
### 필요한 라이브러리 임폴트

from konlpy.tag import Okt, Mecab

# 형태소 분석기 객체 생성
okt = Okt()
mecab = Mecab()
```

```python
### Okt 형태소 분석기 실습

#실습 문장 생성 
text = "한글 형태소 분석기로 테스트를 해보았답니다. 설치 및 동작이 잘 됩니다 :)) "

# 문장 --> 형태소 추출 및 분할(토큰화)--> morphs(stem, norm)

morphs_okt = okt.morphs(text)
morphs_stem_okt = okt.morphs(text, stem = True)

print(morphs_okt)
print(morphs_stem_okt)

splited = text.split()
print(splited)

tokenized_text = word_tokenize(text)
print(tokenized_text)
```

```python
### 형태소와 품사를 동시에 추출 --> 품사 tagging

pos_okt = okt.pos(text)
pos_stem_okt = okt.pos(text, stem = True)

print(pos_okt)
print(pos_stem_okt)
```

```python
### Mecab 형태소 분석기 실습 

#실습 문장 생성 
text = "한글 형태소 분석기로 테스트를 해보았답니다. 설치 및 동작이 잘 됩니다 :)) "

#문장으로부터 형태소 추출 --> morphs()함수 사용 -> 형태소로 분할

morphs_mecab = mecab.morphs(text)

print(morphs_mecab)
print(morphs_okt)
```

```python
pos_mecab = mecab.pos(text)

print(pos_mecab)

print(pos_okt)
```

```python
### 문장에서 명사, 동사, 형용사만 추출

# 품사가 태깅된 데이터 형태 확인

data = pos_stem_okt[0] 
morpheme = data[0] #형태소 추출
pos = data[1] #품사 추출

print(data)
print(morpheme)
print(pos)

# 추출 결과를 저장할 리스트 생성
okt_morphs = []
mecab_morphs = []

#okt 형태소 분석기로 명사,동사, 형용사 추출
for data in pos_stem_okt:
    if 'Noun' in data or 'Verb' in data or 'Adjective' in data:
        okt_morphs.append(data[0])

# Mecab 형태소 분석기를 이용하여 명사, 동사, 형용사 추출

for data in pos_mecab:
    if "NNG" in data[1].split('+') or "NNP" in data[1].split('+') or "VV" in data[1].split('+') or "VA" in data[1].split('+'):
        mecab_morphs.append(data[0])

#결과 확인하기
print(mecab_morphs)
print(okt_morphs)
```

```python
### 한글 텍스트 데이터 다운로드
#필요한 라이브러리 임폴트
from konlpy.corpus import kolaw

#한국 법률 말뭉치 다운로드

law_text = kolaw.open('constitution.txt').read()


print(type(law_text))

print(len(law_text))
```

```python
### 전체 텍스트 데이터 확인

print(law_text)
```

```python
# 필요한 라이브러리 임폴트 
from konlpy.tag import Mecab

#형태소 분석기 객체 생성
mecab = Mecab()

#명사만 추출 --> 토큰화 
nouns_mecab = mecab.nouns(law_text)

#결과 확인하기
print(nouns_mecab)
print(len(nouns_mecab))
```

```python
# 필요한 라이브러리 임폴트 
from konlpy.tag import Okt

#형태소 분석기 객체 생성
okt = Okt()

#명사만 추출 --> 토큰화 
nouns_okt = okt.nouns(law_text)

#결과 확인하기
print(nouns_okt)
print(len(nouns_okt))
```

```python
### 명사로된 형태소의 빈도수 분석

from collections import Counter

counters = Counter(nouns_mecab)

print(counters)
```

```python
### 불용어 제거 --> 형태소 길이가 1보다 큰 형태소만 추출
#리스트 컴프리헨션
nouns = [noun for noun in nouns_mecab if len(noun) >1]

#함수
# nouns = []
# for noun in nouns_mecab:
#     if len(noun) > 1:
#         nouns.append(noun)

print(nouns)
print(len(nouns))
```

```python
### 길이가 1인 불용어가 제거된 품사가 명사인 형태소의 빈도수 분석
counts = Counter(nouns)
print(counts)
```

```python
from wordcloud.wordcloud import WordCloud

import matplotlib.pyplot as plt

font= '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

# WordCloud 객체 생성
wordcloud = WordCloud(
    font_path=font,
    width=1000,
    height= 500,
    max_font_size = 200,
    max_words=50,
    background_color = 'white'
)

# generate_from_frequencies() 함수 사용 --> wordcloud 이미지생성

wordcloud.generate_from_frequencies(counts)

#이미지 저장
wordcloud.to_file('contitution.png')


#이미지 출력
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
